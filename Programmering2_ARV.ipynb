## 2) Exempel → Uppgift
**Exempel 2 – `super()` i `__init__`**

```python
class Djur:
    def __init__(self, namn):
        self.namn = namn

class Katt(Djur):
    def __init__(self, namn, färg):
        super().__init__(namn)  # kalla basklassens konstruktor
        self.färg = färg

k = Katt("Misse", "svart")
print(k.namn, k.färg)
```

**Uppgift 3 – `info()`**
Lägg till en metod `info()` i `Fordon` som **returnerar** en sträng med märke och år.
Överskugga `info()` i `Bil` och lägg till extra info (t.ex. modell). Visa skillnaden.

```python
# TODO Uppgift 3
class Fordon:
    def __init__(self, märke, år):
        self.märke = märke
        self.år = år # första classen fordon med märke och år

    def info(self):
        return f" {self.märke}, {self.år}" #en till funktion i cladd fordon som returnerar tildeade variabler

class Bil(Fordon): #ny class
    def __init__(self, märke, år, modell): #som lägger till variablen modell
        super().__init__(märke, år) #och ersätter märke och år
        self.modell = modell

    def info(self):
        base_info = super().info() #ersätter infon
        return f"{base_info}, {self.modell}" #returnar nya informationen och modellen

f = Fordon("Toyota", 2000) #värderna
b = Bil("Volvo", 2022, "V60") #andra variablerna

print(f.info()) #och printar
print(b.info())
```

## 4) Exempel → Uppgift
**Exempel 4 – `super()` i överskuggad metod**

```python
class Video:
    def info(self):
        return "Allmän video"

class Film(Video):
    def info(self):
        bas = super().info()
        return bas + " – film"

print(Film().info())
```

**Uppgift 4 – `super().info()`**
Använd `super().info()` i `Bil.info()` för att återanvända basklassens text och lägga till bil-specifika detaljer.

```python
# TODO Uppgift 4
class Bil:
    def __init__(self, märke,modell): #Basklass med märke och modell
        self.märke = märke
        self.modell = modell

    def info(self):
        return f"{self.märke} {self.modell}" #returnerar string med variablerna

class Fordon(Bil): #subklass
    def __init__(self, märke, modell):#Samma som basklass
        self.märke = märke
        self.modell = modell

    def info(self): #en till funktion som ersätter bariablerna med super
        bas = super().info()
        return bas +  " 2015" #returnerar de nya variablerna och lägger till årtal

b = Bil("Volvo", "V60") # ger värderna och printar dem
f = Fordon("Toyota", "yaris")
print(b.info())
print(f.info())
```

**Exempel 5 – `isinstance` & `issubclass`**

```python
class A: pass
class B(A): pass

b = B()
print(isinstance(b, A))     # True
print(issubclass(B, A))     # True
```
