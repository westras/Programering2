**Uppgift 3 ‚Äì `info()`**
L√§gg till en metod `info()` i `Fordon` som **returnerar** en str√§ng med m√§rke och √•r.

```python
# TODO Uppgift 3
class Fordon:
    def __init__(self, m√§rke, √•r):
        self.m√§rke = m√§rke
        self.√•r = √•r # f√∂rsta classen fordon med m√§rke och √•r

    def info(self):
        return f" {self.m√§rke}, {self.√•r}" #en till funktion i cladd fordon som returnerar tildeade variabler

class Bil(Fordon): #ny class
    def __init__(self, m√§rke, √•r, modell): #som l√§gger till variablen modell
        super().__init__(m√§rke, √•r) #och ers√§tter m√§rke och √•r
        self.modell = modell

    def info(self):
        base_info = super().info() #ers√§tter infon
        return f"{base_info}, {self.modell}" #returnar nya informationen och modellen

f = Fordon("Toyota", 2000) #v√§rderna
b = Bil("Volvo", 2022, "V60") #andra variablerna

print(f.info()) #och printar
print(b.info())
```

**Uppgift 4 ‚Äì `super().info()`**
Anv√§nd `super().info()` i `Bil.info()` f√∂r att √•teranv√§nda basklassens text och l√§gga till bil-specifika detaljer.

```python
# TODO Uppgift 4
class Bil:
    def __init__(self, m√§rke,modell): #Basklass med m√§rke och modell
        self.m√§rke = m√§rke
        self.modell = modell

    def info(self):
        return f"{self.m√§rke} {self.modell}" #returnerar string med variablerna

class Fordon(Bil): #subklass
    def __init__(self, m√§rke, modell):#Samma som basklass
        self.m√§rke = m√§rke
        self.modell = modell

    def info(self): #en till funktion som ers√§tter bariablerna med super
        bas = super().info()
        return bas +  " 2015" #returnerar de nya variablerna och l√§gger till √•rtal

b = Bil("Volvo", "V60") # ger v√§rderna och printar dem
f = Fordon("Toyota", "yaris")
print(b.info())
print(f.info())
```
