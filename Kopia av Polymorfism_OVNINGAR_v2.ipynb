## Uppgift 3

Skapa en klass `Printer` med metoden `skriv(text, gånger=1)` som skriver text flera gånger. Testa polymorfism genom att skapa en ny klass som har samma metod men skriver texten i VERSALER.

```python
# TODO Uppgift 3
class Printer(): #basklass
  def skriv(self, text, gånger=1):# en metod som skriver ut en text 1 gång
    for i in range(0, gånger):
      print(text)
class VERSALER(Printer):#subklass som använder samma metod som basklass
  def skriv(self, text, gånger=1):
    for i in range(0, gånger):
      print(text.upper())#men skriver ut texten i upper//versaler istället för med små bokstäver

Printer().skriv("hej")#sen printar vi båda metoderna
VERSALER().skriv("hej")
```

## Uppgift 5

Använd en abstrakt klass `Figur` med en metod `area()`. Implementera `Rektangel` och `Cirkel`. Testa polymorfism genom att loopa över en lista av figurer och anropa `area()`.

```python
# TODO Uppgift 5
from abc import ABC, abstractmethod #börjar med att importera abstrakt

class Figur(ABC):#sen gör vi en abstract basklass
  @abstractmethod
  def area(self):
    pass

class Rektangel(Figur): #sen gör vi en subklass för rektangel
  def __init__(self, bredd, höjd):
    self.bredd = bredd
    self.höjd = höjd# där vi har höjd och bredd
  def area(self):
    return self.bredd * self.höjd# sen använder vi area formeln för rektanel med area metoden so mvi skapade i basklassen


class Cirkel(Figur): #vi gör samma med cirkel
  def __init__(self, radie):
    self.radie = radie
  def area(self):
    return self.radie ** 2 * 3.14#samma metod men annan formel alltså radio*radie//radie*2 *3.14

figurer = [Rektangel(5, 10), Cirkel(5)] #vi gör en lista med figurerna

for figur in figurer:
  print(figur.area())#och printar dem
```

## Uppgift 6

Skapa en funktion `visa_info(obj)` som skriver ut resultatet av `obj.info()`. Skapa två olika klasser (`Film`, `Bok`) som båda har en `info()`-metod. Testa med polymorfism.

```python
# TODO Uppgift 6
class information():# en basklass
  def visa_info(self, obj):#basklassen har en metod som visar information
    print(obj.info())
  def info(self):#info metoden som används i subklasserna men används i alla pga polymorfism
    pass

class Film(information): #subklass
  def __init__(self, Titel, regissör):#som har en init för att hantera ny information
    self.Titel = Titel
    self.regissör = regissör

  def info(self): #sen läggs informationen om filmen in i info metoden
    return f"filmens titel är {self.Titel} och regissören är {self.regissör}" #som returnerar en sträng

class Bok(information): #en till subklass
  def __init__(self, Titel, Författare):#som också har en init för att hantera ny information
    self.Titel = Titel
    self.Författare = Författare

  def info(self):#en så används samma metod alltså info för tredje gången
    return f"boken är {self.Titel} och författare är {self.Författare}"

f = Film("Interstellar", "Christopher Nolan") # vi ger information till init
b = Bok("Harry Potter", "J.K. Rowling")

obj = [f, b] #lägger objekten i ne lista
info_printer = information() #tillkallar information via info printer
for objekt in obj:# gör en loop där vi tillkallar visa info på objekten
  info_printer.visa_info(objekt)
```

## Uppgift 9

Skapa en abstrakt klass `Transportmedel` med en metod `framåt()`. Gör subklasser `Bil`, `Cykel`, `Buss`. Testa polymorfism i en lista.

```python
# TODO Uppgift 9
from abc import ABC, abstractmethod

class Transportmedel(ABC):#en abstrakt klass med metoden frammåt och avstånd
  @abstractmethod
  def framåt(self):#metod framåt
    pass

  @abstractmethod
  def avstånd(self): #avstånd metod
    pass

class Bil(Transportmedel): #bil subklass
  def __init__(self, tankstorlek): #har en init där vi lägger till informationen om tank storleken
    self.tankstorlek = tankstorlek
  def framåt(self):
    return "Bilen kör framåt!"#har samma metoder som basklassen men override:as

  def avstånd(self):
    return f"Bilen kan köra ungefär {self.tankstorlek * 10} km med sin tank." #säger att den drar ca 1L/milen

class Cykel(Transportmedel):
  def framåt(self):
    return "Cykeln rör sig framåt!" #har en frammåt metod som override:ar basklassen

  def avstånd(self): #avstånd utan beräkning då cykeln är oerhört kraftfull
    return "Cykeln kan cykla oändligt länge"

class Buss(Transportmedel): #buss subklass
  def __init__(self, tankstorlek):
    self.tankstorlek = tankstorlek #också med tankstorlek

  def framåt(self):
    return "Bussen åker framåt!"

  def avstånd(self):
    return f"Bussen kan åka ungefär {self.tankstorlek * 8} km med sin tank." #säger att den bara drar 0.8L/milen

transportmedel_lista = [Bil(50), Cykel(), Buss(200)] #sen så lägger vi dem i en lista med deras tankstorlekar

for fordon in transportmedel_lista: #och printar listan med både avstånd och framåt metoderna
  print(fordon.framåt())
  print(fordon.avstånd())
```

## Uppgift 10

Skapa en funktion `beräkna_bonus(anställd)` som anropar en metod `bonus()`. Gör klasser `Chef` och `Utvecklare` som returnerar olika bonus. Testa polymorfism.

```python
# TODO Uppgift 10
class Anställd():
  def bonus(self):
    return 0 # alla anställda för 0 bonus vanligtvis

class Chef(Anställd):
  def bonus(self):
    return 1 # chefen är korrupt och får mindre bonus

class Utvecklare(Anställd):
  def bonus(self):
    return 5000 # bonus for a Utvecklare

class Anställd_rasmus(Anställd):
  def bonus(self):
    return 19900 # Exemple bonus for Rasmus

def beräkna_bonus(anställd):
  print(f"Bonus för {anställd.__class__.__name__}: {anställd.bonus()} kr") #printar det på ett fint sätt

Anställda = [Chef(), Utvecklare(), Anställd_rasmus()] #lägger alla i en lista
for anställd in Anställda:
  beräkna_bonus(anställd)
```
